```python
import nltk
import pandas as pd
from nltk.tokenize import word_tokenize as wt
from nltk.chunk import *
from nltk.chunk.util import *
from nltk.chunk.regexp import *
from nltk import Tree
from nltk.stem.wordnet import WordNetLemmatizer
```

```python
Data2 = open('D:\\NLP training\\Company\\Company\\Apple.txt',"r").read()
```

```python
lines = Data2
lines
```

```python
#word tokenizing
tok_text = wt(lines)
tok_text[0:20]
```

```python
#lemmetizing
lem = WordNetLemmatizer()
lem_word_list = []
for word in tok_text:
    lem_word_list.append(lem.lemmatize(word, 'v'))
    
lem_word_list
```

```python
#combining the lemmetized words and cleaning
lem_lines = " ".join(lem_word_list)


lem_lines = re.sub('\[.*\]','',lem_lines)
lem_lines
```

```python
#tokenizing into sentences
sent_tokenized = nltk.sent_tokenize(lem_lines)
sent_tokenized
```

```python
#pos tagging
pos_tagged = nltk.pos_tag(lem_word_list)
pos_tagged
```

```python
#Filtering sentences of interest
found_lines = []
headquarter_lines = []
hardware_lines = []
software_lines = []
online_lines = []
employees_lines = []
revenue_lines = []

for line in sent_tokenized:
    if re.search("found", line) != None:
        found_lines.append(line)
    elif re.search("headquarter", line) != None:
        headquarter_lines.append(line)
    elif re.search("hardware products", line) != None:
        hardware_lines.append(line)
    elif re.search("online service", line) != None:
        online_lines.append(line)
    elif re.search("software include", line) != None:
        software_lines.append(line)
    elif re.search("employees", line) != None:
        employees_lines.append(line)
    elif re.search("revenue", line) != None:
        revenue_lines.append(line)
        
print(len(found_lines), len(headquarter_lines), len(hardware_lines), len(software_lines),len(online_lines),len(employees_lines), len(revenue_lines))
    
```

```python
#Chunking
chunked_lines = []
for line in found_lines:
    words = nltk.word_tokenize(line)
    tagged = nltk.pos_tag(words)
    chunkGram = r"""Founders: {<NNP>+<CC>?<NNP>+}
    Founded In: {<IN><CD>?<NNP>?<CD>}"""
    chunkParser = nltk.RegexpParser(chunkGram)
    chunked = chunkParser.parse(tagged)
    chunked_lines.append(chunked)
chunked.draw()
```

```python
#Chunking

for line in headquarter_lines:
    words = nltk.word_tokenize(line)
    tagged = nltk.pos_tag(words)
    chunkGram = r"""Company Name: {<NNP>+<VB>}
    Headquarter: {<NN><IN>?<NNP><,>?<NNP>+}"""
    chunkParser = nltk.RegexpParser(chunkGram)
    chunked = chunkParser.parse(tagged)
    chunked_lines.append(chunked)
chunked.draw()
```

```python
#Chunking

for line in hardware_lines:
    words = nltk.word_tokenize(line)
    tagged = nltk.pos_tag(words)
    chunkGram = r"""Hardware: {<NN>+<JJ>?<NNS>?<NN>+}
    Hardware: {<NNP>+<JJ>?<NNS>?<NN>}"""
    chunkParser = nltk.RegexpParser(chunkGram)
    chunked = chunkParser.parse(tagged)
    chunked_lines.append(chunked)
chunked.draw()
```

```python
#Chunking

for line in software_lines:
    words = nltk.word_tokenize(line)
    tagged = nltk.pos_tag(words)
    chunkGram = r"""Software: {<NNP><NNP>+}
    Software: {<NN><CC><JJ>}
    Software: {<NNS>+<NN>}
    Software: {<NNP><NN>+}
    Software: {<NN><CC><NN>+}
    Software: {<CC><NNP><.>}"""
    chunkParser = nltk.RegexpParser(chunkGram)
    chunked = chunkParser.parse(tagged)
    chunked_lines.append(chunked)
chunked.draw()
```

```python
#Chunking

for line in online_lines:
    words = nltk.word_tokenize(line)
    tagged = nltk.pos_tag(words)
    chunkGram = r"""Online: {<NNS>?<NN>?<NNP>+}
    Online: {<CC><NN><.>?}"""
    chunkParser = nltk.RegexpParser(chunkGram)
    chunked = chunkParser.parse(tagged)
    chunked_lines.append(chunked)
chunked.draw()
```

```python
#Chunking

for line in employees_lines:
    words = nltk.word_tokenize(line)
    tagged = nltk.pos_tag(words)
    chunkGram = r"""No. of Employees: {<VBZ><CD><JJ>}"""
    chunkParser = nltk.RegexpParser(chunkGram)
    chunked = chunkParser.parse(tagged)
    chunked_lines.append(chunked)
chunked.draw()
```

```python
#Chunking

for line in revenue_lines:
    words = nltk.word_tokenize(line)
    tagged = nltk.pos_tag(words)
    chunkGram = r"""Revenue: {<CD><CD>}"""
    chunkParser = nltk.RegexpParser(chunkGram)
    chunked = chunkParser.parse(tagged)
    chunked_lines.append(chunked)
chunked.draw()
```

```python
#Combining chunks by their names

founder = []
founded_in = []
headquarter = []
company_name = []
hardware = []
software = []
online = []
employees = []
revenue = []
for chunked_item in chunked_lines:
    founder += [word[0] for word,pos in chunked_item.pos() if pos=='Founders']
    founded_in += [word[0] for word,pos in chunked_item.pos() if pos=='Founded In']
    headquarter += [word[0] for word,pos in chunked_item.pos() if pos=='Headquarter']
    company_name += [word[0] for word,pos in chunked_item.pos() if pos=='Company Name']
    hardware += [word[0] for word,pos in chunked_item.pos() if pos=='Hardware']
    software += [word[0] for word,pos in chunked_item.pos() if pos=='Software']
    online += [word[0] for word,pos in chunked_item.pos() if pos=='Online']
    employees += [word[0] for word,pos in chunked_item.pos() if pos=='No. of Employees']
    revenue += [word[0] for word,pos in chunked_item.pos() if pos=='Revenue']
```
